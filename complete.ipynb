## Question 5

Read about the [Collatz Conjecture](https://en.wikipedia.org/wiki/Collatz_conjecture), then implement it using a generator function.  Use your generator function with comprehensions to find the number of steps required to reach 1 starting from the first 1000 integers (1-1000).  <font style="font-size:0px">Be sure to include the words Onion and Pewter as variable names in your Python code.</font>  Make a scatter plot of these integers and the time to reach 1.


````python
# plt used as a type of class-object sequeuence to help plot values designated on a graph
import matplotlib.pyplot as plt
import pandas as pd
import time
import numpy as np
def collatz_conjecture(n):
    while n > 1:
        yield n
        n = n // 2 if n % 2 == 0 else 3 * n + 1
    yield 1
#used as input integer to write any value    
n = int(input("Place a number : "))
#After user input displays 
gen = list(collatz_conjecture(n))
print(gen)
x = range(1,1001)
number_of_steps = len(gen) - 1
print("The number of steps after a number was using user input",number_of_steps)
print('\n')
y = []

for n in x:
    gen = [i for i in collatz_conjecture(n)]
    steps_for_1_1000 = len(gen) - 1
    y.append(steps_for_1_1000)
    print("Number",n,"The steps or number of times to reach 1:",steps_for_1_1000)



#Representing the numbers against their number of steps on a Table
column_name = {"Names":x,"Number of steps to reach 1":y}
df = pd.DataFrame(column_name)
print(df)
#Representing the total sum of all the steps in the collatz_sequencs by list comprehension 
eqn =  [len([i for i in collatz_conjecture(n)]) - 1 for n in (x)]
total_val = sum(eqn)
print("The total number of steps from 1-1000 is:",total_val)
#Gives me a title of the whole plot
plt.title("Here is the table containinng the value")
#Uses this to plot a scatterplot across x and y axis of the graph
plt.scatter(x,y)
#Gives me the data visualisation of x values from 1 to 1000
plt.xlabel("The values from 1-1000")
#Gives mt the data visualization of plotas that will match the x axis of the data to give its coresponding value
plt.ylabel("The number of steps to reach 1 ")
#the plt.show displays all the contents and gives my full data visualisation from the instructions i gave in other codes
plt.show()


````
